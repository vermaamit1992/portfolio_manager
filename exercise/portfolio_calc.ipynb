```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
from pandas_datareader import data as wb
```

```python
stock= ['DMART.NS','RELIANCE.NS','TCS.NS','HDB']
new=pd.DataFrame()
for t in stock:
    new[t]= wb.DataReader(t, data_source='yahoo',start='2019-1-1')['Close']
```

```python
new.to_csv('C:/Users/AMIT/Desktop/py/first.csv')
```

```python
returns=(new/new.shift(1))-1
returns.head()
```

```python
weights=np.array([0.25,0.25,0.25,0.25])
```

```python
np.dot(returns,weights)
```

```python
ret_a=returns.mean()*250
ret_a
```

```python
np.dot(ret_a,weights)
```

```python
pfolio1= str(round(np.dot(ret_a,weights),3)*100)+'%'
print (pfolio1)
```
