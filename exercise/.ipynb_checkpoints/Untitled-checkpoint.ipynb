```python
#price forecasting

import numpy as np
import pandas as pd
from pandas_datareader import data as wb
import matplotlib.pyplot as plt
from scipy.stats import norm
%matplotlib inline
```

```python
stocks =['TCS.NS']
new_data = pd.DataFrame()
for t in stocks:
    new_data[t] = wb.DataReader(t, data_source='yahoo', start='2010-1-1')['Close']
```

```python
log_return = np.log(1 + new_data.pct_change())
log_return.head()
```

```python
log_return.plot(figsize=(8,5))
```

```python
u = log_return.mean()
u
```

```python
var =  log_return.var()
var
```

```python
# drift = mean - 0.5*var
drift = u - 0.5 * var
drift
```

```python
# normalization
(new_data/new_data.iloc[0]*100).plot(figsize=(8,5))
plt.show()
```

```python
log_returns=np.log(new_data/new_data.shift(1))
```

```python
log_returns.mean() * 250
```

```python
log_returns.cov() * 250
```

```python
log_returns.corr()
```

```python
num_stocks= len(stocks)
```

```python
weights= np.random.random(num_stocks)
weights /= np.sum(weights)
weights
```

```python
#expected portfolio return

np.sum(weights * log_returns.mean() * 250)
```

```python
#expected portfolio variance

np.dot(weights.T, np.dot(log_returns.cov() * 250, weights))
```

```python
#expected portfolio volatility
np.sqrt(np.dot(weights.T, np.dot(log_returns.cov() * 250, weights)))
```

```python
pfolio_returns = []
pfolio_volatilities = []
for x in range (1000):
    weights = np.random.random(num_stocks)
    weights /= np.sum(weights)
    pfolio_returns.append(np.sum(weights * log_returns.mean()) * 250)
    pfolio_volatilities.append(np.sqrt(np.dot(weights.T, np.dot(log_returns.cov() * 250, weights))))
pfolio_returns
```

```python
pfolio_volatilities
```

```python
pfolio_returns = np.array(pfolio_returns)
pfolio_volatilities=np.array(pfolio_volatilities)
```

```python
portfolios = pd.DataFrame({'Returns':pfolio_returns, 'Volatility': pfolio_volatilities})
```

```python
portfolios.plot(x='Volatility', y='Returns', kind= 'scatter', figsize=(8,6))
plt.xlabel('Expected Volatility')
plt.ylabel('Expected Returns')
plt.show()
```
