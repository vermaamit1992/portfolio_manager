```python
import numpy as np
import pandas as pd
from pandas_datareader import data as wb
import matplotlib.pyplot as plt
```

```python
ind = ['^BSESN','^NSEI','^NSEBANK']
new = pd.DataFrame()
for i in ind:
    new[i] = wb.DataReader(i, data_source='yahoo', start = '2019-1-1')['Adj Close']
    
```

```python
new.info()
```

```python
## normalization
(new/new.iloc[0]*100).plot(figsize=(15,5))
plt.show()
```

```python
ind_return_d = (new/new.shift(1))-1

ind_return_d.head()
```

```python
ind_return_a = ind_return_d.mean()*250
ind_return_a
```
